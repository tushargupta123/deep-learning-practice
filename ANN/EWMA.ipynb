# Estimate weighted moving average

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Date'],df['Temp'],color='red')
```

```python
x1 = df['Temp'].ewm(alpha=0.9).mean()
```

```python
x1
```

```python
df['ewma'] = x1
```

```python
df.head()
```

```python
plt.scatter(df['Date'],df['Temp'],color='red')
plt.plot(df['Date'],df['ewma'],color='black',label='alpha=0.9')
plt.legend()
plt.show()
```
