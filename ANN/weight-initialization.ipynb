```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
X,y = make_moons(100,noise=0.25,random_state=2)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

# Xavier / Glorot Weight Initialization

can only be applied on tanh  or sigmoid activation function

## Normal function

initialize weights with random from 0 to 1 and divide it by sqrt(1/n) where n is the number of nodes in previous layer

## Uniform function

initialize weights with random from 0 to 1 and divide it by sqrt(6/n+m) where n (fan in ) is the number of nodes in previous layer , m (fan  out) is the number of node in next layer 

```python
model = Sequential()

model.add(Dense(10,activation='tanh',input_dim=2))
model.add(Dense(10,activation='tanh'))
model.add(Dense(10,activation='tanh'))
model.add(Dense(10,activation='tanh'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.get_weights()
```

```python
initial_weights = model.get_weights()
```

```python
initial_weights[0] = np.random.randn(2,10)*np.sqrt(1/2)
initial_weights[1] = np.zeros(model.get_weights()[1].shape)
initial_weights[2] = np.random.randn(10,10)*np.sqrt(1/10)
initial_weights[3] = np.zeros(model.get_weights()[3].shape)
initial_weights[4] = np.random.randn(10,10)*np.sqrt(1/10)
initial_weights[5] = np.zeros(model.get_weights()[5].shape)
initial_weights[6] = np.random.randn(10,10)*np.sqrt(1/10)
initial_weights[7] = np.zeros(model.get_weights()[7].shape)
initial_weights[8] = np.random.randn(10,1)*np.sqrt(1/10)
initial_weights[9] = np.zeros(model.get_weights()[9].shape)
```

```python
model.set_weights(initial_weights)
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history = model.fit(X,y,epochs=100,validation_split=0.2)
```

```python
model.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'),clf=model,legend=2)
```

# He Initilization

used in case of relu activation

## Normal function

initialize weights with random from 0 to 1 and divide it by sqrt(2/n) where n is the number of nodes in previous layer

## Uniform function

initialize weights with random from 0 to 1 and divide it by sqrt(6/n) where n (fan in ) is the number of nodes in previous layer

```python
model = Sequential()

model.add(Dense(10,activation='relu',input_dim=2,kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_uniform'))
model.add(Dense(10,activation='tanh',kernel_initializer='glorot_normal'))
model.add(Dense(10,activation='tanh'))  # by default kernal initializer is set to glorot_uniform
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history = model.fit(X,y,epochs=100,validation_split=0.2)
```

```python
plot_decision_regions(X,y.astype('int'), clf=model, legend=2)
```
