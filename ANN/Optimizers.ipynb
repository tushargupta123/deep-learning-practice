# Momentum and Nesterov Momentum

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,BatchNormalization
```

```python
X,y = make_moons(100,noise=0.25,random_state=2)
```

```python
model = Sequential()

model.add(Dense(2,activation='relu',input_dim=2))
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
sgd = tensorflow.keras.optimizers.SGD(learning_rate=0.01,momentum=0,nesterov=False)
```

```python
model.compile(loss='binary_crossentropy',optimizer=sgd,metrics=['accuracy'])
history1 = model.fit(X,y,epochs=100,validation_split=0.2)
```

```python
momentum = tensorflow.keras.optimizers.SGD(learning_rate=0.01,momentum=0.9,nesterov=False)
```

```python
model.compile(loss='binary_crossentropy',optimizer=momentum,metrics=['accuracy'])
history2 = model.fit(X,y,epochs=100,validation_split=0.2)
```

```python
nesterov = tensorflow.keras.optimizers.SGD(learning_rate=0.01,momentum=0.9,nesterov=True)
```

```python
model.compile(loss='binary_crossentropy',optimizer=nesterov,metrics=['accuracy'])
history3 = model.fit(X,y,epochs=100,validation_split=0.2)
```

```python
plt.plot(history1.history['accuracy'],label='sgd')
plt.plot(history2.history['accuracy'],label='momentum')
plt.plot(history3.history['accuracy'],label='nesterov')
plt.legend()
plt.show()
```
